```python
# Libraries, Read Out Dataset

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
dataset = pd.read_csv('Saudi Arabian GP 2023.csv')
dataset.head(20)
```

```python
# X = Drivers ID, y = Final Position, Z = Constructors ID, H = Grid

X = dataset.iloc[:-2, 2:-6].values
y = dataset.iloc[:-2, :-8].values
Z = dataset.iloc[:-2, 4:-4].values
H = dataset.iloc[:-2, 8:].values
```

```python
# Machine Learning Regression Grid vs Final Position

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(H, y)
```

```python
# Saudi Arabian GP 23 Driver Position

plt.figure(figsize=(15,10))
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.scatter(X, y, color = 'blue')
plt.title('Saudi Arabian GP 23 Driver Position')
plt.xlabel('Driver ID')
plt.ylabel('Position')
```

```python
# Saudi Arabian GP 23 Constructor Position

plt.figure(figsize=(15,10))
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.scatter(Z, y, color = 'blue')
plt.title('Saudi Arabian GP 23 Constructor Positions')
plt.xlabel('Constructor ID')
plt.ylabel('Position')
```

```python
# Saudi Arabian GP 23 Driver Points

f, ax = plt.subplots(figsize = (15, 10))
sns.barplot(x='points', y='driver', data=dataset)
plt.title('Saudi Arabian GP 23 Driver Points')
plt.xlabel('GP Points')
plt.ylabel('Drivers')
```

```python
# Saudi Arabian GP 23 Contructor Points

f, ax = plt.subplots(figsize=(15,10))
sns.barplot(x='points', y='constructor', data=dataset)
plt.title('Saudi Arabian GP 23 Constructor Points')
plt.xlabel('GP Points')
plt.ylabel('Constructors')
```

```python
# Saudi Arabian GP 23 Grid VS Finish Points

plt.figure(figsize=(15, 10))
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.scatter(H, y, color = 'red')
plt.plot(H, regressor.predict(H), color = 'blue')
plt.title('Grid VS Finish')
plt.xlabel('Start Position')
plt.ylabel('Finish Position')
```
